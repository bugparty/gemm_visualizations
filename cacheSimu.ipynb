````python
KB=1024
cache_size = 32*KB
A_base = 0x10000
B_base = 0x20000
C_base = 0x30000
D_base = 0x40000
E_base = 0x50000
type_size = 8 #bytes
def strAddr(base,i):
    addr = base + i * type_size
    block_offset = addr & 0b111111 
    set_index = (addr >> 6) & 0b111111
    tag = addr >> 12
    
    return "{} {} {}".format(hex(tag)[2:].rjust(6,'0'), bin(set_index)[2:].rjust(6,'0'),bin(block_offset)[2:].rjust(6,'0'))
for i in range(0,20):
    print('{}[{}] {}'.format('A',i,strAddr(A_base,i)))
    print('{}[{}] {}'.format('B',i,strAddr(B_base,i)))
    print('{}[{}] {}'.format('C',i,strAddr(C_base,i)))
    print('{}[{}] {}'.format('D',i,strAddr(D_base,i)))
    print('{}[{}] {}'.format('E',i,strAddr(E_base,i)))
    


````
